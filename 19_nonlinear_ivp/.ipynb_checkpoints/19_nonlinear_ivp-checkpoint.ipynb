# Initial Value Problems (continued)

## Predator-Prey Models (Chaos Theory)

Predator-prey models were developed independently in the early part of the twentieth
century by the Italian mathematician Vito Volterra and the American biologist Alfred
Lotka. These equations are commonly called Lotka-Volterra equations. The simplest version is the following pairs of ODEs:

$\frac{dx}{dt}=ax-bxy$

$\frac{dy}{dt}=-cy+dxy$

where x and y = the number of prey and predators, respectively, a = the prey growth rate, c = the predator death rate, and b and d = the rates characterizing the effect of the predator-prey interactions on the prey death and the predator growth, respectively.

```matlab
function yp=predprey(t,y,a,b,c,d)
    % predator-prey model (Lotka-Volterra equations)
    yp=zeros(1,2);
    x=y(1);
    y=y(2);
    yp(1)=a.*x-b.*x.*y;
    yp(2)=-c*y+d*x.*y;
end
```

````matlab
h=0.0625;tspan=[0 40];y0=[2 1];
a=1.2;b=0.6;c=0.8;d=0.3;
````

````matlab
predprey(1,[20 1],a,b,c,d)
````

````matlab
a=1.2;b=0.6;c=0.8;d=0.3;
h=0.001;
[t y] = eulode(@predprey,tspan,y0,h,a,b,c,d);

plot(t,y(:,1),t,y(:,2),'--')
legend('prey','predator');
title('(a) Euler time plot')
````

````matlab
plot(y(:,1),y(:,2))
title('(b) Euler phase plane plot')
xlabel('prey population')
ylabel('pred population')
````

````matlab
a=10;b=0.6;c=0.8;d=0.3;
tspan=[0 10];
[t y] = rk4sys(@predprey,tspan,y0,h,a,b,c,d);
plot(t,y(:,1),t,y(:,2),'--')
title('(c) RK4 time plot')
````

````matlab
plot(y(:,1),y(:,2))
title('(d) RK4 phase plane plot')
````

# Systems of ODE's:
## The Coriolis Effect

System of ODE's defined as:
```matlab
  function ddr = coriolis_ode(t,r,R,L)
    g=9.81; % acceleration due to gravity m/s^2
    l=10; % 10 m long cable
    we=2*pi/23.934/3600; % rotation of Earth (each day is 23.934 hours long)
    ddr=zeros(4,1); % initialize ddr

    ddr(1) = r(2); % x North(+) South (-)
    ddr(2) = 2*we*r(4).*sin(L)-g/l*r(1); % dx/dt 
    ddr(3) = r(4); % y West (+) East (-)
    ddr(4) = -2*we*(r(2).*sin(L))-g/l*r(3); % dy/dt
  end
```

First, we will solve for the path of a neutrally-buoyant object with a 100-N force pointing North. It's path starts at 41.8084$^o$ lattitude (Storrs, CT)

````matlab
north_coriolis(41.8084,1)
````

````matlab
coriolis(41.8084)
````
